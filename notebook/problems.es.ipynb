## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random

import numpy as np
def lanzamiento_dados():
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    return dado1 + dado2

num_lanzamientos = 1000
contador_exito = 0

for _ in range(num_lanzamientos):
    suma_dados = lanzamiento_dados()
 
    if suma_dados > 7 or suma_dados % 2 == 0:
        contador_exito += 1


probabilidad_estimada = contador_exito / num_lanzamientos
print("La probabilidad estimada es:", probabilidad_estimada)
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
def sacarbolas(ball_box):
    bolas_sacadas = [random.choice(list(ball_box.values())) for _ in range(5)]
    return bolas_sacadas


num_simulaciones = 1000
tres_blancas_2rojas = 0
todas_del_mismo_color = 0

for _ in range(num_simulaciones):
    bolas_sacadas = sacarbolas(ball_box)
    

    if bolas_sacadas.count("White") == 3 and bolas_sacadas.count("Red") == 2:
        tres_blancas_2rojas += 1
    

    if len(set(bolas_sacadas)) == 1:
        todas_del_mismo_color += 1

#### re estudiar esta ultima parte por que me ayudaron 
probabilidad_3blancas_2rojas = tres_blancas_2rojas / num_simulaciones
probabilidad_todas_del_mismo_color = todas_del_mismo_color / num_simulaciones

print("3 bolas blancas y 2 rojas:", probabilidad_3blancas_2rojas)
print(" mismo color:", probabilidad_todas_del_mismo_color)
```
